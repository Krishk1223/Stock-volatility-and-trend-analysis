## General Conclusions based on data (not stock specific):

- Prices are non stationary (log returns are a better approach here)
- Log returns fluctuate around zero (no persistent drift, implying not able to meaningfully predict mean).
- Rolling Volatility varies with time (constant volatility cannot be assumed).
- ACF and PACF of log returns are close to zero across lags (markets around 2020 were hectic due to pandemic so major fluctuations there), therefore the returns are not easily predictable (white noise like). No signal structure makes it less likely to choose ARIMA models.
- ACF and PACF of squared log returns show positive autocorrelation with decay implying that variance is time-dependent (heteroskedasticity implied, GARCH models may be justified).
- Returns bear fatter tails (see QQ plots) implying log returns are not normally distributed (underestimating extremes). Maybe see other distributions like student-t distribution.
- Correlation is present between assets (AMD and Google are prime examples with a 0.53 correlation), must model joint behaviour (covariance matrix needed).

As such given returns remain approximately uncorrelated but volatility is time varying and that assets bear correlation the choice of model + simulation will be as follows:

- GARCH (Generalised AutoRegressive Conditional Heteroskedasticity) models (possibly GARCH(1,1 where variance at time depends on day before's squared residual + predicted variance))
- Monte Carlo Simulations (with both normal and student t distribution)
