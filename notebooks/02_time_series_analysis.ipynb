### AMD plots

Below we see 7 plots for Advanced Micro Devices (AMD) showing the following:

- Close price over Time
- Log Returns over Time (with rolling mean as well as rolling standard deviation with a window of 21 trading days)
- Annualised (over 252 trading days within a year) Volatility Over Time
- A histogram showcasing the distribution of the log returns overlaid with a normal distribution (to help see how skewed the tails of data are).
- ACF (autocorrelation function) of log returns data (for ARIMA justification through predicability of returns). Looking for direct + indirect effect of lags.
- ACF (autocorrelation function) of squared log returns data (for GARCH justification through relation to volatility). Looking for direct + indirect effect of lags.
- PACF (Partial Autocorrelation function) of log returns. Looking for direct effect of lags.
- PACF (Partial Autocorrelation function) of squared log returns (once again for GARCH justification). Looking for direct effect of lags.

```python
stock_choice = 'AMD'
print(f"Showing time series analysis plots for {stock_choice}...")
plt.show(stock_show_plots(features_with_stats, stock_choice))
```

```python
print(f"Showing correlation functions for {stock_choice}...")
plt.show(stock_show_acf_pacf(features_with_stats, stock_choice))
```
